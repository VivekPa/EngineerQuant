### Computing Historical Betas

```python
import statsmodels.api as sm
from scipy.stats import gaussian_kde
from scipy import integrate
from statsmodels.stats.stattools import jarque_bera
```

```python
sp_500_ret = np.log(sp_500/sp_500.shift(1)).dropna()
spy_ret = np.log(spy/spy.shift(1)).dropna()

sp_500_ret_pre = sp_500_ret.loc['2017-01-01':'2020-01-01']
spy_ret_pre = spy_ret.loc['2017-01-01':'2020-01-01']
```

```python
amzn = sp_500_ret['AMZN'].loc['2019-01-01':'2020-01-01']
x = sm.add_constant(spy_ret.loc[amzn.index])
y = amzn

model = sm.OLS(y, x)
results = model.fit()
results.summary()
```

```python
plt.scatter(spy_ret.loc[amzn.index], y)
```

```python
spcov = pd.concat([spy_ret.loc[amzn.index], amzn], axis=1, sort=False).cov()

amzn_beta = spcov.loc['AMZN', 'SPY']/spcov.loc['SPY', 'SPY']

amzn_beta
```

```python
for i in spy_ret_pre.index == sp_500_ret_pre.index:
    if not i:
        print('Not aligned')
```

```python
beta_df = pd.Series(index=sp_500_ret_pre.columns)

for i in sp_500_ret_pre.columns:
    current = sp_500_ret[i]
    y = current
    X = sm.add_constant(spy_ret)
    
    model = sm.OLS(y, X)
    results = model.fit()
    beta_df.loc[i] = results.params[1]
    
beta_df.head()
```

### Historical Sector Betas 

```python
sp_const.index = sp_const['Symbol']
r_sp_const = sp_const.loc[sp_500.columns]
sectors = r_sp_const['Sector'].unique()

sector_beta = pd.Series(index=sectors)

for i in sectors:
    curr_sector = r_sp_const.loc[r_sp_const['Sector'] == i]['Symbol']
    curr_sector_sp = sp_500_ret_pre.loc[:, curr_sector]
    
    sec_beta = []
    for j in curr_sector:
        y = curr_sector_sp[j]
        X = sm.add_constant(spy_ret_pre)
        
        model = sm.OLS(y, X)
        results = model.fit()
        curr_beta = results.params[1]
        sec_beta.append(curr_beta)
    mean_beta = np.mean(sec_beta)
    sector_beta.loc[i] = mean_beta
```

```python
sector_beta_pre = sector_beta
sector_beta_pre
```

```python
plt.figure(figsize=(12,6))

plt.title('Sector Beta (2017-2020)')
plt.bar(sector_beta_pre.index, sector_beta_pre)
plt.plot(sector_beta_pre.index, np.ones(len(sector_beta_pre)), 'g-.')
plt.xticks(rotation=45)
plt.ylabel(r'$ \beta $')
plt.show()
```

```python
begin = 600
date_set = spy_ret.index[begin:]
rs_beta = []

for n in range(len(date_set)):
    start = spy_ret.index[begin+n-100]
    curr_date = spy_ret.index[begin+n]

    sp_500_ret_dur = sp_500_ret.loc[start:curr_date]
    spy_ret_dur = spy_ret.loc[start:curr_date]

    sp_const.index = sp_const['Symbol']
    r_sp_const = sp_const.loc[sp_500.columns]
    sectors = r_sp_const['Sector'].unique()

    sector_beta = pd.Series(index=sectors)

    rs_sector = r_sp_const.loc[r_sp_const['Sector'] == 'Real Estate']['Symbol']
    rs_sector_sp = sp_500_ret_dur.loc[:, rs_sector]

    sec_beta = []
    for i in rs_sector:
        y = rs_sector_sp[i]
        X = sm.add_constant(spy_ret_dur)

        model = sm.OLS(y, X)
        results = model.fit()
        curr_beta = results.params[1]
        sec_beta.append(curr_beta)

    mean_beta = np.mean(sec_beta)
    rs_beta.append(mean_beta)
    
rs_beta = pd.Series(rs_beta, index=date_set)
rs_beta
```

```python
plt.figure(figsize=(12,6))

plt.title('3M Rolling Beta for Real Estate')
plt.plot(rs_beta)
plt.xticks(rotation=45)
plt.ylabel(r'$ \beta $')
plt.show()
```
